```python
#Clase donde generare la lista de libros
class catalogo:
    """
    En esta clase se definira la lista de libros para poder
    llevar a cabo un orden, asi como poder llevar un conteo de 
    los libros existentes disponibles
    """

    libro=[] #Lista vacia donde se guardaran los libros

    def _init_(self, libro=[]):
        self.libro=libro

    def agregar(self, li): #Aqui se agregan los libros a la lista
        self.libro.append(li)

    def mostrar(self): #Con este metodo mostramos los datos
        for li in self.libro:
            print(li)
```
